* stanovit cetnosti jednotlivych znaku v retezci
    * prochazet retezec znak po znaku
         * muzete bud zjistit kod znaku pomoci funkce ord, a v poli s indexem kodu znaku pricist 1
         * pouzit slovnik, kde klicem je znak a hodnotou je jeho cetnost


* vykreslit sloucovy graf cetnosti znaku

```python
string = "Hello world"
cetnosti={} # klic je znak a hodnota bude cetnost
for i in string:
    cetnosti[i]=cetnosti.get(i,0)+1
print(cetnosti)
```

```python
import matplotlib.pyplot as plt
plt.bar(cetnosti.keys(), cetnosti.values())
plt.show()
```

```python
string = "Hello world"
cetnosti=[0]*256 # klic je znak a hodnota bude cetnost
znaky = [0]*256
for i in range(256):
    znaky[i]=chr(i)

for i in string:
    kod = ord(i)
    cetnosti[kod]=cetnosti[kod]+1
print(cetnosti)

x=[]
y=[]
for i in range(256):
    if cetnosti[i] !=0:
        x.append(chr(i))
        y.append(cetnosti[i])

plt.bar(x, y)
```

* stahnete si soubor z https://mech.fsv.cvut.cz/~bp/download/PythonCourse/world_cities.txt
* bude vhodne pouzit modul csv, ktery usnadbnuje praci s csv soubory
* stanovit cetnosti mest v jednotlivych zemich, ktera maji vice nez 10M obyvatel

```python
import csv

f = open('Data/world_cities.txt')
csv_reader = csv.reader(f)
cetnosti={}
for line in csv_reader:
    country=line[4]
    if line[9].isnumeric(): 
        population=int(line[9])
        if (population > 10000000):
            cetnosti[country]=cetnosti.get(country,0)+1
print(cetnosti)
```

```python
plt.bar(cetnosti.keys(), cetnosti.values())
plt.xticks(rotation=90)
```
